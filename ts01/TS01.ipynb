1. Obtener la función transferencia (módulo, fase y diagrama de polos y ceros).
Planteo la corriente que circula por R1 y R2, y la que circula por C y R3:

$$
I_1=I_{R1}=I_{R2}=\frac{V_1-V_{in}}{R_1}=\frac{V_{in}-V_2}{R_2}
$$

$$
I_2=I_{R3}=I_C=\frac{V_1-V_{in}}{\frac{1}{SC}}=\frac{V_{in}}{R_3}
$$

Despejando $V_{in}$ de $I_1$

$$
\frac{V_1}{R_1}-\frac{V_{in}}{R_1}=\frac{V_{in}}{R_2}-\frac{V_2}{R_2}
$$

$$
\frac{V_1}{R_1}+\frac{V_2}{R_2}=V_{in}(\frac{1}{R_1}+\frac{1}{R_2})
$$

$$
V_{in}\frac{R_1+R_2}{R_1R_2}=\frac{V_1R_2+V_2R_1}{R_1R_2}
$$

$$
V_{in}=\frac{V_1R_2+V_2R_1}{R_1+R_2}
$$

Despejando $V_{in}$ de $I_2$

$$
V_1SC=V_{in}(SC+\frac{1}{R_3})
$$

$$
V_{in}=\frac{V_1SC}{SCR_3C+1}
$$

Igualando ambas expresiones y desarrollando, se llega a la siguiente expresión:

$$
\frac{V_2}{V_1}=\frac{SR_1R_3C-R_2}{SR_1R_3C+R_1}=\frac{S-\frac{R_2}{R_1R_3C}}{S+\frac{1}{R_3C}}
$$

Para obtener las expresiones de módulo y fase de la transferencia, reemplazo $s$ por $jw$:

$$
T(w)=\frac{jw-\frac{R_2}{R_1R_3C}}{jw+\frac{1}{R_3C}}
$$

$$
T(w)=\frac{\sqrt{w^2+\frac{R_2^2}{R_1^2R_3^2C^2}}}{\sqrt{w^2+\frac{1}{R_3^2C^2}}}\frac{e^{j\arctan(\frac{-wR_1R_2C}{R_2})}}{e^{j\arctan(wR_3C)}}
$$

$$
T(w)=\frac{\sqrt{w^2+\frac{R_2^2}{R_1^2R_3^2C^2}}}{\sqrt{w^2+\frac{1}{R_3^2C^2}}}\frac{e^{j\arctan(\frac{-wR_1R_2C}{R_2})}}{e^{j\arctan(wR_3C)}}
$$

$$
|T(w)|=\sqrt{\frac{w^2+\frac{R_2^2}{R_1^2R_3^2C^2}}{w^2+\frac{1}{R_3^2C^2}}}
$$

$$
\gamma(T)=-\arctan(\frac{wR_1R_2C}{R_2})-\arctan(wR_3C)
$$
